```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Example 1 - A simple cost function
## $$f(x)=x^2+x+1$$


```python
def f(x):
    return x**2+x+1
```

```python
#Make Data
```

```python
x_1 = np.linspace(start=-3, stop=3, num = 100)
```

```python
x_1
```

```python
#Plot
plt.xlim([-3, 3])
plt.ylim(0, 8)
plt.xlabel('X', fontsize= 16)
plt.ylabel('f(x)', fontsize = 16)
plt.plot(x_1, f(x_1))
plt.show()
```

## Slope and Derivatives


```python
def df(x):
    return 2*x + 1
```

```python
#Plot function side by side
plt.figure(figsize=[15, 5])
# 1 Chart: Cost Function
plt.subplot(1, 2, 1)
plt.xlim([-3, 3])
plt.ylim(0, 8)
plt.title('Cost Function', fontsize=17)
plt.xlabel('X', fontsize= 16)
plt.ylabel('f(x)', fontsize = 16)
plt.plot(x_1, f(x_1), color='blue', linewidth=3)
# 2 Chart: Derivative
plt.subplot(1, 2, 2)
plt.xlim([-2, 3])
plt.ylim(-3, 6)
plt.grid()
plt.title('Slope of Cost Function', fontsize=17)
plt.xlabel('X', fontsize= 16)
plt.ylabel('df(x)', fontsize = 16)
plt.plot(x_1, df(x_1), color='skyblue', linewidth=5)
plt.show()
```

## Python Loops & Gradient Descent

```python
# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.01
prec = 0.0001

x_list = [new_x]
slope_list = [df(new_x)]


for i in range(1000):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier*gradient
    
    
    step_size = abs(new_x - previous_x)
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    #print(step_size)
    
    if step_size < prec:
        print("Loop ran for:", i)
        break
    
print("Local Minimum occurs at: ", new_x)
print("Slope of the curve at this point is :", df(new_x))
print("f(x) value or cost or loss at this point is:",f(new_x))
    
```

```python
#Superimpose gradient descent values
plt.figure(figsize=[15, 5])
# 1 Chart: Cost Function
plt.subplot(1, 2, 1)
plt.xlim([-3, 3])
plt.ylim(0, 8)
plt.title('Cost Function', fontsize=17)
plt.xlabel('X', fontsize= 16)
plt.ylabel('f(x)', fontsize = 16)
plt.plot(x_1, f(x_1), color='blue', linewidth=4)
values = np.array(x_list)
plt.scatter(x_list, f(values), color='red', s=100, alpha=0.4)
# 2 Chart: Derivative
plt.subplot(1, 2, 2)
plt.xlim([-2, 3])
plt.ylim(-3, 6)
plt.grid()
plt.title('Slope of Cost Function', fontsize=17)
plt.xlabel('X', fontsize= 16)
plt.ylabel('df(x)', fontsize = 16)
plt.plot(x_1, df(x_1), color='skyblue', linewidth=5)
plt.scatter(x_list, slope_list, color='red', s=100, alpha=0.4)


plt.show()
```

# Example 2 - Multiple Minima vs Initial Guess Advanced Functions

## $$g(x) = x^4 - 4x^2 + 5$$

```python
# Make some data
x_2 = np.linspace(-2, 2, 1000)

#g(x) and dg(x) function
def g(x):
    return x**4-4*x**2+5
def dg(x):
    return 4*x**3-8*x
```

```python
#Plot function side by side
plt.figure(figsize=[15, 5])
# 1 Chart: Cost Function
plt.subplot(1, 2, 1)
plt.xlim([-2, 2])
plt.ylim(0.5, 5.5)
plt.title('Cost Function', fontsize=17)
plt.xlabel('X', fontsize= 16)
plt.ylabel('g(x)', fontsize = 16)
plt.plot(x_2, g(x_2), color='blue', linewidth=3)
# 2 Chart: Derivative
plt.subplot(1, 2, 2)
plt.xlim([-2, 2])
plt.ylim(-6, 8)
plt.grid()
plt.title('Slope of Cost Function', fontsize=17)
plt.xlabel('X', fontsize= 16)
plt.ylabel('dg(x)', fontsize = 16)
plt.plot(x_2, dg(x_2), color='skyblue', linewidth=5)
plt.show()
```
